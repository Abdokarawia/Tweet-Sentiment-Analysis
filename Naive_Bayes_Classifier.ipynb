```python
# Downloading the stopwards
nltk.download('stopwords')
```

```python
# Arabic stopwards
stop_words = stopwords.words('arabic')
arabic_diacritics=re.compile("""
                             ّ    | # Shadda
                             َ    | # Fatha
                             ً    | # Tanwin Fath
                             ُ    | # Damma
                             ٌ    | # Tanwin Damm
                             ِ    | # Kasra
                             ٍ    | # Tanwin Kasr
                             ْ    | # Sukun
                         """, re.VERBOSE)
```

```python
stop_words
```

```python
text_cleaning_re = '@\S+|https?:\S+|http?:\S.\S|'
```

# Tokenization

```python
from nltk.stem.isri import ISRIStemmer
stemmer = ISRIStemmer()
```

```python
#c = df['Tweet'].apply(nltk.word_tokenize)
```

```python
#filtered_text=[]
#for word in df['Tweet']:
 # if word in stop_words:
  #  filtered_text.append(word)
```

```python
def preprocess(Text, stem = False):
    #print(Text)
    tokens =[]
    for word in Text.split():
        word= re.sub(text_cleaning_re,'',word)   # replace text_cleaning_re with space 
        word=re.sub(arabic_diacritics,'',word)
        if word not in stop_words:
            if stem:
                tokens.append(stemmer.stem(word))
            else:
                tokens.append(word)
                
        print(tokens)
    return ' '.join(tokens)
```

```python
df['Tweet_after_clean']= df['Tweet'].apply(preprocess)
```

# Binary encoding
 relation between dicument, term..deals with the vocab.

```python
from sklearn.feature_extraction.text import CountVectorizer  # CountVectorizer : get a list depent on n gram that i want.
vec = CountVectorizer(binary = True)
features = vec.fit_transform(df['Tweet_after_clean'])
# fit : conver text to dic to get vocabs, key is vocab and the value is its reapet in the dataset.
# transform : take raw text and start to make 2d matrix to show the relation between every vocab and the sentiment.
# sklearn ignore char that equal 1.
```

```python
pd.DataFrame.sparse.from_spmatrix(features, columns=sorted(vec.vocabulary_.keys()))
# make virtualization to the values inside the matrix.
```

# Naive Bayes Classifier (Multinomial)

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features , df['Sentiment'], test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
nb_classifier=model.fit(x_train, y_train)
```

```python
predicted = model.predict(x_test)   # x_test : array of values.
print(predicted)
```

```python
print(x_test)
```

# Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test , predicted)
```

# confusion matrix

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, predicted))
```

```python
# define a function to predict the sentiment of a new text message
def predict_sentiment(text):
    # preprocess the text message
    text_clean = preprocess(text)
    # transform the cleaned text into a bag-of-words matrix
    text_bow = vec.transform([text_clean])
    # use the trained Naive Bayes classifier to predict the sentiment
    sentiment = nb_classifier.predict(text_bow)[0]
    return sentiment
```
